```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

t1_path = "COMP9517_25T1_Lab1_Images/Task1.jpg"
t2_path = "COMP9517_25T1_Lab1_Images/Task2.jpg"
t3_path = "COMP9517_25T1_Lab1_Images/Task3.jpg"

t1 = cv2.imread(t1_path,cv2.IMREAD_GRAYSCALE)
t2 = cv2.imread(t2_path,cv2.IMREAD_GRAYSCALE)
t3 = cv2.imread(t3_path,cv2.IMREAD_GRAYSCALE)
```

Task 1

```python
def gamma_correction(image, gamma):
    normalized = image / 255.0
    corrected = np.power(normalized, gamma)

    return np.uint8(corrected * 255)

inver_t1 = 255 - t1

gamma = 1.5
gamma_t1 = gamma_correction(inver_t1, gamma)

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.imshow(t1, cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(gamma_t1, cmap="gray")
plt.title("Result")
plt.axis("off")
plt.show()
```

Task 2

```python
def median_filter(image, kernel):
    return cv2.medianBlur(image, kernel)

t2_med = median_filter(t2, 3)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(t2, cmap="gray")
plt.title("Orginal")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(t2_med, cmap="gray")
plt.title("Result")
plt.axis("off")
plt.show()
```

I choose median filter to process this photo. Guassian is the best choice for retaining small objects. Both the mean filter and the median filter can reduce the noise. But median filter is better than mean filter, because it preserves edges while removing nosie.

Task 3

```python
t3_gauss = cv2.GaussianBlur(t3, (3, 3), 0)
h_fre = np.clip(t3 - t3_gauss, 0, 255)
a = 6
res = np.clip(t3 + a * h_fre, 0, 255)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(t3, cmap="gray")
plt.axis("off")
plt.title("Original")

plt.subplot(1, 2, 2)
plt.imshow(res, cmap="gray")
plt.title("Result")
plt.axis("off")
plt.show()
```

I decide to use unsharp masking method. Laplacian are more commonly used for edge detection, but unsharp masking is better for image sharpening.
